```python
#ประกาศการเรียกใช้งาน
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

```

```python
#Lab01
sin_wave = thinkdsp.SinSignal(freq=7010,amp=10,offset=0)
sin_wave.plot()
thinkplot.config(xlabel = 'Time(s)')
```

```python
#lab02
#trianglesignal(freq)
#duration=signal.period*n
signal=thinkdsp.TriangleSignal(10)
duration=signal.period*10
segment=signal.make_wave(duration,framerate=10000)
segment.plot()

snwave=signal.make_wave(duration,framerate=10000)
snwave.apodize()
snwave.make_audio()

spectrum=snwave.make_spectrum()
spectrum.plot()
```

```python
signal=thinkdsp.SquareSignal(10)
duration=signal.period*10
segment=signal.make_wave(duration,framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05,1.05],legend=False)
sqwave=signal.make_wave(duration,framerate=10000)
sqwave.apodize()
sqwave.make_audio()
```
